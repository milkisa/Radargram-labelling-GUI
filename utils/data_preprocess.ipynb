```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

rs_image= pd.read_csv('E:/important/phd/project/new_project/segment_anything/sharad/dataset/sharad_rs.csv', header = None).to_numpy().reshape(-1,912,64)
rs_sharad= pd.read_csv('E:/important/phd/project/new_project/segment_anything/sharad/dataset/sharad_gt.csv', header = None).to_numpy().reshape(-1,912,64)
```

```python
image_trans = rs_image.transpose(1, 0, 2)
image= image_trans.reshape(912,-1)
label_trans = rs_sharad.transpose(1, 0, 2)
label= label_trans.reshape(912,-1)
```

```python
plt.imsave('image.png', image)
plt.imsave('label.png', label)
```
